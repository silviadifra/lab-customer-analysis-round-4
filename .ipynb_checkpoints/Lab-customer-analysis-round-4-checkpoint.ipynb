```python
#Show a plot of the total number of responses.
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline

fig, ax = plt.subplots()
ax = sns.histplot(file1.response, color='tab:green')
plt.title('response')
plt.show()
```

```python
#Show a plot of the response rate by the sales channel.

fig, ax = plt.subplots(figsize=(10,5))
sns.histplot(x='sales channel', hue='response', data=file1, multiple='dodge')
plt.show()
```

```python
#Show a plot of the response rate by the total claim amount.
fig, ax = plt.subplots(figsize=(10,5))
sns.histplot(x='total claim amount', hue='response', data=file1, bins = 50)
plt.show()
```

```python
#Show a plot of the response rate by the income.
fig, ax = plt.subplots(figsize=(10,5))
sns.histplot(x='income', hue='response', data=file1, bins = 50)
plt.show()
```
